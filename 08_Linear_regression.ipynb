```
#import tensorflow as tf
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
x_train = [1, 2, 3]
```

```
y_train = [2+1 +0.1,  4+1 -0.2, 6+1 +0.15]
```

```
plt.plot(x_train, y_train,'.-')
plt.xlabel('x')
plt.ylabel('y')
```

y=2x +1
y=wx+b



```
w0 = 6;
b0 = 8;
```

```
w = tf.Variable(w0*tf.ones([1]), name='weight')
b = tf.Variable(b0*tf.ones([1]), name='bias')
```

```
 hypothesis = x_train * w+b
```

```
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
```

```
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
#optimizer = tf.compat.vl.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

```
sess = tf.Session()
```

```
sess.run(tf.global_variables_initializer())
```

```
nb_epoch = 2000
```

```
for i in range(100):
  if i % 20 ==0:
   print(i)
```

```
vw = []
vb = []
vcost = []
```

```
for step in range(nb_epoch+1):
    sess.run(train)

    wl = sess.run(w)[0]
    bl = sess.run(b)[0]
    costl= sess.run(cost)

    vw.append(wl)
    vb.append(bl)
    vcost.append(vcost)

    if step % 100 == 0:
        print(step, costl, wl, bl)
```

```
wl = sess.run(w)[0]
bl = sess.run(b)[0]
strl = 'y={:.3}x + {:.3}'.format(wl,bl)
print(wl, bl)
print(strl)
```

```
plt.figure(1)
plt.plot(x_train, y_train,'o')
x1 = np.min(x_train)-1, np.max(x_train)+1
y1 = np.dot(wl,x1) + bl
plt.plot(x1, y1) 
plt.grid()
plt.title(strl)
```
