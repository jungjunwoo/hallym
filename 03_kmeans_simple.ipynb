```
import matplotlib.pyplot as plt
```

```
from sklearn.cluster import KMeans
```

```
x = [25,34,22,27,33,33,31,22,35,34,67,54,57,43,50,57,59,52,65,47,49,48,35,33,44,45,]
y = [79,51,53,78,59,74,73,57,69,75,51,32,40,47,53,36,35,58,59,50,25,20,14,12,20,5,]
```

```
import numpy as np
```

```
data = np.array([x, y])
```

```
data.shape
```

```
data2 = data.transpose()
```

```
data2.shape
```

```
plt.plot(x, y, '.')
plt.xlabel('X')
plt.ylabel('Y')
```

```
kmeans = KMeans(n_clusters=3)
```

```
%%time
kmeans.fit(data2)
```

```
centroids = kmeans.cluster_centers_
```

```
print(centroids)
```

```
plt.plot(x, y, '.')
plt.plot(centroids[:, 0], centroids[:, 1], 'o')
plt.xlabel('X')
plt.ylabel('Y')
```
